```python
import pandas as pd

# Cargar los archivos .parquet
transacciones = pd.read_parquet('transacciones.parquet', engine='fastparquet')
clientes = pd.read_parquet('clientes.parquet', engine='fastparquet')
productos = pd.read_parquet('productos.parquet', engine='fastparquet')

# Vista previa de los datos
display(transacciones.head())
display(clientes.head())
display(productos.head())
```

```python
#Se genera un df uniendo los 3 dfs
df_data_full = (
    transacciones
    .merge(clientes, on='customer_id', how='left')   # Une transacciones con clientes
    .merge(productos, on='product_id', how='left')   # Une el resultado con productos
)

# Vista r√°pida de las primeras filas
display(df_data_full.head())
```

```python
#Se revisan duplicados
num_duplicadas = df_data_full.duplicated().sum()
print(f"Filas duplicadas encontradas: {num_duplicadas:,}")

#Se eliminan duplicados
df_data_full = df_data_full.drop_duplicates()

print(f"Filas despu√©s de eliminar duplicados: {df_data_full.shape[0]:,}")
```

```python
df_data_full['items_bin'] = (df_data_full['items'] > 0).astype(int)
```

```python
#Se genera un df con todas las combinaciones posibles de customer_id y product_id
clientes_ids = clientes['customer_id'].drop_duplicates().reset_index(drop=True)
productos_ids = productos['product_id'].drop_duplicates().reset_index(drop=True)
mi = pd.MultiIndex.from_product(
    [clientes_ids.values, productos_ids.values],
    names=['customer_id', 'product_id']
)
df_cliente_producto = mi.to_frame(index=False)

df_cliente_producto = (
    df_cliente_producto
    .merge(clientes, on='customer_id', how='left')
    .merge(productos, on='product_id', how='left')
)
```

```python
#Se proyecta el df
df_cliente_producto
```

```python

# ----- 1) Asegurar tipos y columnas necesarias -----
transacciones['purchase_date'] = pd.to_datetime(transacciones['purchase_date'], errors='coerce')
# columnas que traeremos de transacciones
tx_cols = ['customer_id', 'product_id', 'order_id', 'purchase_date', 'items']

# ----- 2) Enriquecer transacciones con atributos de cliente y producto (para que coincidan con df_cliente_producto) -----
trans_full = (
    transacciones[tx_cols]
    .merge(clientes, on='customer_id', how='left', suffixes=('', '_cli'))
    .merge(productos, on='product_id', how='left', suffixes=('', '_prod'))
)

# ----- 3) Preparar df_cliente_producto base -----
base = df_cliente_producto.copy()

# A√±adir columnas de transacci√≥n iniciales con los valores requeridos (NaNs y 0)
base['order_id'] = pd.NA
base['purchase_date'] = pd.NaT
base['items'] = 0

# ----- 4) Separar la "primera" transacci√≥n por par (customer_id, product_id) y las restantes -----
# Ordenamos por fecha para tomar la transacci√≥n m√°s antigua o la que prefieras (aqu√≠: la primera por fecha ascendente)
trans_sorted = trans_full.sort_values('purchase_date').reset_index(drop=True)

# Primera transacci√≥n por combinaci√≥n
mask_duplicated = trans_sorted.duplicated(subset=['customer_id', 'product_id'], keep='first')
first_tx = trans_sorted[~mask_duplicated].copy()    # primera tx por par
rest_tx = trans_sorted[mask_duplicated].copy()      # el resto de tx (si hay m√°s de una por par)

# ----- 5) Merge: llenar filas base con la primera transacci√≥n si existe -----
# Solo traemos las columnas de tx que queremos llenar
first_tx_trim = first_tx[['customer_id', 'product_id', 'order_id', 'purchase_date', 'items']]

# Merge base con first_tx_trim (left merge para mantener todas las combinaciones base)
base_merged = base.merge(first_tx_trim,
                         on=['customer_id', 'product_id'],
                         how='left',
                         suffixes=('', '_txfirst'))

# Si la transacci√≥n existe, usar los valores de la transacci√≥n; si no, mantener NaN/0 preexistentes
# Reemplazar las columnas order_id, purchase_date, items con las de la transacci√≥n cuando no sean null
base_merged['order_id'] = base_merged['order_id_txfirst'].combine_first(base_merged['order_id'])
base_merged['purchase_date'] = base_merged['purchase_date_txfirst'].combine_first(base_merged['purchase_date'])
base_merged['items'] = base_merged['items_txfirst'].combine_first(base_merged['items'])

# Eliminar columnas temporales de merge
base_merged = base_merged.drop(columns=[c for c in base_merged.columns if c.endswith('_txfirst')])

# ----- 6) Para las transacciones "restantes", crear filas nuevas replicando atributos -----
# rest_tx ya contiene los atributos de cliente/producto por la fusi√≥n previa (trans_full)
# Necesitamos seleccionar columnas en el mismo orden/nombres que base_merged

# Determinar columnas no transaccionales (atributos del cliente/producto) que existen en base_merged
# Consideramos que base_merged contiene al menos: customer_id, product_id y dem√°s atributos
attrs = [c for c in base_merged.columns if c not in ['order_id', 'purchase_date', 'items']]

# Construir las filas adicionales a partir de rest_tx:
# Primero seleccionar de rest_tx los atributos del cliente/producto y las columnas de transacci√≥n
# rest_tx tiene columnas: customer_id, product_id, order_id, purchase_date, items, + atributos cliente/producto + atributos producto
# Alinearemos con los nombres de base_merged: attrs + ['order_id','purchase_date','items']

# Si rest_tx no contiene exactamente las mismas columnas de atributos, nos aseguramos de seleccionar las que s√≠ existen
available_attrs = [c for c in attrs if c in rest_tx.columns]
rows_extra = rest_tx[ available_attrs + ['order_id', 'purchase_date', 'items'] ].copy()

# Si faltan atributos en rows_extra (porque rest_tx no conten√≠a alguna columna), a√±adimos columnas vac√≠as
for c in attrs:
    if c not in rows_extra.columns:
        rows_extra[c] = pd.NA

# Reordenamos columnas para que coincidan con base_merged
rows_extra = rows_extra[attrs + ['order_id', 'purchase_date', 'items']]

# ----- 7) Concatenar base_merged (una fila por combinaci√≥n) con las filas extra -----
df_cliente_producto_final = pd.concat([base_merged[ attrs + ['order_id', 'purchase_date', 'items'] ], rows_extra], ignore_index=True, sort=False)

# ----- 8) Opcional: ordenar y reset index -----
df_cliente_producto_final = df_cliente_producto_final.reset_index(drop=True)

# ----- 9) Informar resultados -----
print("Filas base (combinaciones cliente-producto):", len(base))
print("Primeras transacciones aplicadas (llenadas en base):", len(first_tx))
print("Transacciones adicionales que generaron nuevas filas:", len(rest_tx))
print("Filas finales en df_cliente_producto_final:", len(df_cliente_producto_final))

# Mostrar primeras filas
df_cliente_producto_final.head(10)
```

En esta secci√≥n, se deben construir pipelines para automatizar el preprocesamiento de los datos, lo cual garantizar√° que el flujo de trabajo sea reproducible y eficiente para esta entrega y las futuras. El objetivo es aplicar una serie de transformaciones en un orden definido para asegurar que los datos est√©n listos para los modelos a entrenar. El pipeline final debe incluir las t√©cnicas de pre-procesamiento que se deben aplicar a los distintos datos (seg√∫n lo que consideren necesario para el problema). Por ejemplo:

- **Imputaci√≥n de valores nulos**: Manejo de datos faltantes mediante estrategias adecuadas (media, mediana, moda, interpolaci√≥n, etc.).

- **Transformaciones personalizadas**: Uso de ColumnTransformer para aplicar diferentes transformaciones a columnas espec√≠ficas.

- **Codificaci√≥n de variables categ√≥ricas**: Convertir datos categ√≥ricos a un formato num√©rico adecuado (One-Hot Encoding, Label Encoding, etc.).

- **Discretizaci√≥n de variables**: Conversi√≥n de variables num√©ricas continuas en categor√≠as si son relevantes para el desempe√±o del modelo a entrenar.

- **Estandarizaci√≥n o normalizaci√≥n** : Ajustar la escala de los datos para mejorar el rendimiento de los algoritmos sensibles a la magnitud de las variables.

- **Eliminaci√≥n o transformaci√≥n de valores at√≠picos**: Identificar y tratar con datos outliers para mejorar la robustez del modelo.

- **Nuevas caracter√≠sticas** : Creaci√≥n de variables adicionales que puedan aportar informaci√≥n relevante al modelo.

Cada una de estas transformaciones debe ser justificada en funci√≥n de su relevancia para el problema y los datos, y es importante evaluar su impacto en el rendimiento del modelo. Adem√°s, el pipeline debe ser flexible y modular para poder probar diferentes configuraciones de preprocesamiento.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, KBinsDiscretizer
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin

class OutlierHandler(BaseEstimator, TransformerMixin):
    """
    Clase para tratar outliers y valores inv√°lidos:
    - items < 0 -> se reemplaza por np.nan
    - coordenadas fuera de rango -> se reemplaza por np.nan
    """
    def __init__(self, col_items='items', col_x='X', col_y='Y'):
        self.col_items = col_items
        self.col_x = col_x
        self.col_y = col_y

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()

        if self.col_items in X.columns:
            X.loc[X[self.col_items] < 0, self.col_items] = np.nan

        if self.col_x in X.columns:
            X.loc[(X[self.col_x] < -180) | (X[self.col_x] > 180), self.col_x] = np.nan

        if self.col_y in X.columns:
            X.loc[(X[self.col_y] < -90) | (X[self.col_y] > 90), self.col_y] = np.nan
        return X


# Columnas num√©ricas continuas
num_cols = [ 'num_deliver_per_week', 'num_visit_per_week', 'X', 'Y']

# Columnas categ√≥ricas
cat_cols = ['items_bin','brand', 'category', 'sub_category', 'segment', 'package', 'customer_type', 'region_id', 'zone_id']

# Columna que podr√≠a discretizarse
disc_cols = ['items']


# Pipeline para variables num√©ricas
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pipeline para variables categ√≥ricas
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Pipeline para discretizaci√≥n de variables
disc_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('discretizer', KBinsDiscretizer(n_bins=5, encode='onehot-dense', strategy='quantile'))
])


preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols),
    ('disc', disc_pipeline, disc_cols)
])


full_pipeline = Pipeline([
    ('outlier_handler', OutlierHandler()),   # Transformaciones personalizadas
    ('preprocessor', preprocessor)           # Preprocesamiento est√°ndar
])

full_pipeline.fit(df_train)
training_processed= full_pipeline.fit_transform(df_train)
validation_processed= full_pipeline.fit_transform(df_validation)
test_processed= full_pipeline.fit_transform(df_test)


print(f"Tama√±o del dataset transformado: {training_processed.shape}")
print(f"Tama√±o del dataset transformado: {validation_processed.shape}")
print(f"Tama√±o del dataset transformado: {test_processed.shape}")
```

Principalmente se consideraron datos erroneos como items negativos, (no puedes comprar menos de 0 items) y tambien ubicaciones de latitud y longitud fuera de rango, colocandose como NaN.

En cuanto a el manejo de los NaN se decidio por simplicidad a los datos numericos imputarlos mediante la mediana y los categoricos por la moda, con esto se evita introducir sesgo dentro de los datos y ahorra recursos computacionales comparado con otros metodos mas sofisticados.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler  # (quita KBins de baseline)
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.dummy import DummyClassifier
from sklearn.metrics import classification_report

# --------------------------
# 0) Handler de outliers
# --------------------------
class OutlierHandler(BaseEstimator, TransformerMixin):
    def __init__(self, col_items='items', col_x='X', col_y='Y'):
        self.col_items = col_items
        self.col_x = col_x
        self.col_y = col_y

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Si ya no es DataFrame (p.ej. matriz sparse), no hacer nada
        if not hasattr(X, 'columns'):
            return X

        X = X.copy()
        if self.col_items in X.columns:
            X.loc[X[self.col_items] < 0, self.col_items] = np.nan
        if self.col_x in X.columns:
            X.loc[(X[self.col_x] < -180) | (X[self.col_x] > 180), self.col_x] = np.nan
        if self.col_y in X.columns:
            X.loc[(X[self.col_y] < -90) | (X[self.col_y] > 90), self.col_y] = np.nan
        return X


# --------------------------
# 1) Definir target y listas de columnas SIN fuga
# --------------------------
# Target binario
y_train = df_train['items_bin'].astype(int)
y_val   = df_validation['items_bin'].astype(int)
y_test  = df_test['items_bin'].astype(int)

# Features: NO incluir items_bin (target) ni items (si define el target de esta semana)
num_cols = ['num_deliver_per_week', 'num_visit_per_week', 'X', 'Y']  # continuas
# Ajusta cat_cols a columnas reales disponibles en tus dataframes
cat_cols = ['brand', 'category', 'sub_category', 'segment', 'package',
            'customer_type', 'region_id']  # quit√© items_bin, zone_id

# --------------------------
# 2) Pipelines de preprocesamiento
# --------------------------
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler(with_mean=False))  # with_mean=False para trabajar bien con sparse
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=True))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, [c for c in num_cols if c in df_train.columns]),
    ('cat', cat_pipeline, [c for c in cat_cols if c in df_train.columns]),
], remainder='drop')  # o 'passthrough' si lo necesitas

# --------------------------
# 3) Pipeline completo con clasificador dummy (baseline)
# --------------------------
baseline_pipe = Pipeline([
    ('outliers', OutlierHandler()),
    ('prep', preprocessor),
    ('clf', DummyClassifier(strategy='most_frequent'))
])

# --------------------------
# 4) Fit SOLO en train, transform/predict en val y test
# --------------------------
X_train = df_train.drop(columns=['items_bin'])  # asegurarse de no pasar el target dentro de X
X_val   = df_validation.drop(columns=['items_bin'])
X_test  = df_test.drop(columns=['items_bin'])

baseline_pipe.fit(X_train, y_train)

# Validaci√≥n (opcional, para ver estabilidad)
y_val_pred = baseline_pipe.predict(X_val)
print("VALIDACI√ìN - Baseline DummyClassifier")
print(classification_report(y_val, y_val_pred, digits=4))

# Test (reporte final del baseline)
y_test_pred = baseline_pipe.predict(X_test)
print("TEST - Baseline DummyClassifier")
print(classification_report(y_test, y_test_pred, digits=4))
```

```python
# ===============================================
# Optuna (versi√≥n simple A)
# Modelo: DecisionTreeClassifier
# - Mantiene el preprocesamiento del baseline_pipe
# - Optimiza hiperpar√°metros principales del √°rbol
# - Usa F1-macro como m√©trica
# ===============================================

import optuna
from sklearn.base import clone
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

RANDOM_STATE = 42
SCORING = "f1_macro"
N_SPLITS = 3
N_TRIALS = 15  # puedes ajustar seg√∫n tiempo disponible

def objective(trial):
    # Hiperpar√°metros a optimizar del √°rbol
    max_depth = trial.suggest_int("max_depth", 3, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 50)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 30)
    max_features = trial.suggest_categorical("max_features", [None, "sqrt", "log2"])
    criterion = trial.suggest_categorical("criterion", ["gini", "entropy"])

    # Clonar el pipeline base y setear el clasificador
    pipe = clone(baseline_pipe).set_params(
        clf=DecisionTreeClassifier(
            random_state=RANDOM_STATE,
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_samples_leaf=min_samples_leaf,
            max_features=max_features,
            criterion=criterion
        )
    )

    # Validaci√≥n cruzada
    cv = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=RANDOM_STATE)
    score = cross_val_score(pipe, X_train, y_train, scoring=SCORING, cv=cv).mean()
    return score

# Crear estudio Optuna
study = optuna.create_study(direction="maximize", study_name="DecisionTree_opt")
study.optimize(objective, n_trials=N_TRIALS)

print(" Mejor F1-macro (CV):", study.best_value)
print(" Mejores hiperpar√°metros:")
for k, v in study.best_params.items():
    print(f"  - {k}: {v}")

# Reentrenar con los mejores hiperpar√°metros
best_pipe = clone(baseline_pipe).set_params(
    clf=DecisionTreeClassifier(
        random_state=RANDOM_STATE,
        **study.best_params
    )
)

best_pipe.fit(X_train, y_train)

# Evaluar en VALIDATION
y_pred_val = best_pipe.predict(X_val)
print("\n [DecisionTree + Optuna] Validation set:")
print(classification_report(y_val, y_pred_val, target_names=['items=0','items>0']))
print(" Matriz de confusi√≥n (val):")
print(confusion_matrix(y_val, y_pred_val))

# Evaluar en TEST
y_pred_test = best_pipe.predict(X_test)
print("\n [DecisionTree + Optuna] Test set:")
print(classification_report(y_test, y_pred_test, target_names=['items=0','items>0']))
print(" Matriz de confusi√≥n (test):")
print(confusion_matrix(y_test, y_pred_test))
```

Se decidio usar F1_score dado que creemos que da el mejor resultado para no perder ventas, se asume que se pueden perder ventas por 2 motivos


*   No se le promociona un producto a un cliente que podria comprarlo
*   Se sobrecarga el call center con recomendaciones que en verdad no le interesan a los clientes, haciendo que no de tiempo a darles las recomendaciones que realmente les gustarian a los clientes

F1-Score balancea y cuida que no ocurra ninguna de esas 2 situaciones, en caso de no tener problemas de capacidad se podr√≠a ocupar recall (perder una venta es menos doloroso que dar un aviso de m√°s)

El hiperparametro con mayor impacto fue max depth, consiguiendo el mejor resultado con max_depth=30, aunque no consigue mejorias con respecto al modelo sin optimizacion de hiperparametros dado que ese no tenia limite en max_depth y parece indicar que a mayor max depth mejorres resultados

La verdad el modelo no mejoro despues de la optimizacion de hiperparametros dado que el parametro de mayor importancia es max_depth que ya tenia un valor adecuado anteriormente
